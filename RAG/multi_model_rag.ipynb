```python
from unstructured.partition.pdf import partition_pdf 

raw_pdf = partition_pdf(
    filename = "openai.pdf", 
    strategy = "hi_res", # Hi Resolution Image we want to store. 
    extract_images_in_pdf = True, 
    extract_image_block_types = ["Image", "Table"],
    extract_image_block_to_payload = False, 
    extract_image_block_output_dir = "extracted_data"
)

raw_pdf
```

```python
rpdf = raw_pdf[0:17]
rpdf
```

```python
Header = []
Footer = []
Title = []
NarrativeText = []
Text = []
ListItem = []
Images = []
Table = []

for element in raw_pdf:
    if "unstructured.documents.elements.Header" in str(type(element)):
        Header.append(str(element))
    if "unstructured.documents.elements.Footer" in str(type(element)):
        Footer.append(str(element))
    if "unstructured.documents.elements.Title" in str(type(element)):
        Title.append(str(element))
    if "unstructured.documents.elements.Text" in str(type(element)):
        Text.append(str(element))
    if "unstructured.documents.elements.ListItem" in str(type(element)):
        ListItem.append(str(element))
    if "unstructured.documents.elements.Images" in str(type(element)):
        Images.append(str(element))
    if "unstructured.documents.elements.Table" in str(type(element)):
        Table.append(str(element))
```
